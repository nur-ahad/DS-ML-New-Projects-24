```python
# Importing necessary libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset 

data=pd.read_csv("C://Users//nural//Downloads//WHO-COVID-19-global-data.csv")
```

```python
# Peforming Summary Statistics 

data.describe()
```

```python
# EDA 

# Distribution of Cases by Country 

plt.figure(figsize=(16,8))
sns.countplot(data=data, x ='Country', order =data['Country'].value_counts().index)
plt.title('Distribution of Cases by Country')
plt.xticks(rotation=90)
plt.show()
```

```python
# Total New cases over time 

data['Date_reported'] = pd.to_datetime(data['Date_reported'])

# Plotting total cases over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Date_reported', y='New_cases', estimator='sum')
plt.title('Total New Cases Over Time')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.show()
```

```python
# Total Cumulative Cases Over Time 

data['Date_reported'] = pd.to_datetime(data['Date_reported'])

# Plotting total cases over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Date_reported', y='Cumulative_cases', estimator='sum')
plt.title('Total Cumulative Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Cases')
plt.show()
```

```python
# Total New Cases Across WHO Region 

# Plotting total cases over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='WHO_region', y='New_cases', estimator='sum')
plt.title('Total New Cases Across WHO Region')
plt.xlabel('WHO_region')
plt.ylabel('New Cases')
plt.show()
```

```python
# Total New Cases Across Countries

# Plotting total cases over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Country_code', y='New_cases', estimator='sum')
plt.title('Total New Cases Across Countries')
plt.xlabel('Country')
plt.ylabel('New Cases')
plt.show()
```

```python
# Total New cases over time 

data['Date_reported'] = pd.to_datetime(data['Date_reported'])

# Plotting total cases over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Date_reported', y='New_deaths', estimator='sum')
plt.title('Total New Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('New Deaths')
plt.show()
```

```python
# Total Cumulative Death Cases Over Time 

data['Date_reported'] = pd.to_datetime(data['Date_reported'])

# Plotting total cases over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Date_reported', y='Cumulative_deaths', estimator='sum')
plt.title('Total Cumulative Death Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Death Cases')
plt.show()
```
