```python
# Importing packages 

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset 

df = pd.read_csv("C://Users//nural//Downloads//euro-daily-hist_1999_2022.csv")
```

```python
# Displaying first few rows of the dataset 
df.head()
```

```python
# Checking the data types 

df.dtypes
```

```python
# Generating summary of the Quantitative Features of the Dataset
df.describe()
```

```python
# Generating summary of the Object Features of the Dataset

object_columns = df.select_dtypes(include=['object']).columns
```

```python
# Displayng summary statistics for object columns 

object_summary= df[object_columns].describe()
```

```python
print(object_summary)
```

```python
# Plotting bar blots for Object Features (Using Matplotlib)

object_columns = df.select_dtypes(include=['object']).columns

for column in object_columns:
    plt.figure(figsize=(8,6))
    df[column].value_counts().plot(kind='bar')
    plt.title(f'Bar Plot of Currency')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.show()
```

```python
# Plotting bar blots for Object Features (Using Seaborn)

object_columns = df.select_dtypes(include=['object']).columns

for column in object_columns:
    plt.figure(figsize=(8,6))
    sns.countplot(x=column, data=df, palette='viridis')
    plt.title(f'Count Plot of Currency')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Checking for Missing Values

df.isnull().sum()
```

```python
# Detailed code for missing values 

missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Imputing missing values with mean 

df_filled_mean =df.fillna(df.mean())
```

```python
# Checking for missing values after imputation 

missing_values_after = df_filled_mean.isnull().sum()
print(missing_values_after)
```

```python
# Droping Rows with missing values 

df_dropped = df.dropna()
```

```python
# Checkign before and after shape of the data 
print("Shape before dropping:", df.shape)
print("Shape after dropping:", df_dropped.shape)
```
