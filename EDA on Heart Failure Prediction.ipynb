```python
#Sty# Plot a box plot to show the relationship between numerical and categorical variable
dataset.boxplot(column='age', by='anaemia')
# Set x and y-axis labels
plt.xlabel('anaemia')
plt.ylabel('age')
# Set the title of the plot
plt.title('Relationship between Age and Anaemia')

# Show the plot
plt.show()
```

```python
#Style 5 
# Create a box plot with age on the y-axis and anaemia on the x-axis
plt.boxplot([dataset[dataset['anaemia']==0]['age'], dataset[dataset['anaemia']==1]['age']], labels=['No Anaemia', 'Anaemia'])

# Set the axis labels
plt.xlabel('Anaemia')
plt.ylabel('Age')

# Show the plot
plt.show()
```

```python
#Style 6 
# Create a violin plot with age on the y-axis and anaemia on the x-axis
sns.violinplot(x="anaemia", y="age", data=dataset)

# Set the axis labels
plt.xlabel('Anaemia')
plt.ylabel('Age')

# Show the plot
plt.show()
```

```python
#Style 7 
# Create a swarm plot with age on the y-axis and anaemia on the x-axis
sns.swarmplot(x="anaemia", y="age", hue="anaemia", data=dataset)

# Set the axis labels
plt.xlabel('Anaemia')
plt.ylabel('Age')

# Show the plot
plt.show()
```

```python
# Create a swarm plot with age on the y-axis and anaemia on the x-axis
sns.swarmplot(x="anaemia", y="age", hue="anaemia", data=dataset)

# Set the axis labels
plt.xlabel('Anaemia')
plt.ylabel('Age')

# Add horizontal lines for the mean age of each group
mean_anaemia = dataset[dataset['anaemia'] == 1]['age'].mean()
mean_no_anaemia = dataset[dataset['anaemia'] == 0]['age'].mean()
plt.axhline(mean_anaemia, color='r', linestyle='--', label='Mean Age with Anaemia')
plt.axhline(mean_no_anaemia, color='g', linestyle='--', label='Mean Age without Anaemia')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
#Age with all categorical variables 
# Set the plot style
sns.set(style='ticks')

# Create the subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

# Plot the distribution of age vs. anaemia
sns.histplot(data=dataset, x='age', hue='anaemia', kde=True, ax=axs[0, 0])

# Plot the distribution of age vs. diabetes
sns.histplot(data=dataset, x='age', hue='diabetes', kde=True, ax=axs[0, 1])

# Plot the distribution of age vs. high_blood_pressure
sns.histplot(data=dataset, x='age', hue='high_blood_pressure', kde=True, ax=axs[1, 0])

# Plot the distribution of age vs. sex
sns.histplot(data=dataset, x='age', hue='sex', kde=True, ax=axs[1, 1])

# Plot the distribution of age vs. smoking
sns.histplot(data=dataset, x='age', hue='smoking', kde=True)

# Show the plot
plt.show()
```
